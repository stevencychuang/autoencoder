```python
from time import time
import numpy as np
numSeed = 42
np.random.seed(numSeed)
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt
from keras.layers import Input, Dense, Lambda
from keras.models import Model
from keras import backend as K
from keras import metrics
from keras.datasets import mnist

import tensorflow as tf
from keras.backend.tensorflow_backend import set_session

config = tf.ConfigProto()
config.gpu_options.per_process_gpu_memory_fraction = 0.5
set_session(tf.Session(config=config))

from util import plotScatterDecode, plotProgress, plotCompDecode, plotScatterEncode, addNoise
from dataPreproc import *
import pickle
```

## Declare the parameters

```python
numEpochs = 200
sizeBatch = 32
dimInter1 = 128
dimInter2 = 32
dimInter3 = 8
dimEncode = 2
nameOptim = 'adam'
```

## Load the datasets

```python
with open('./Img_list.pkl', 'rb') as f:
    img_list_all = pickle.load(f)
img_width = 245
img_height = 205
    
# Special downsampling for the videos which are with over 300 frames
img_list_all[0] = img_list_all[0][160:160]
img_list_all[8] = img_list_all[9][110:210]
img_list_all[9] = img_list_all[9][116:216]

# List to numpy array and get corresponding labels
data, prog = list2NumpyArr(img_list_all, img_height, img_width)
data = data[:, 50:150, 60:200] # 100 x 140
labelList, labelSet = getLab(img_list_all)

xTrain = data
xTest = data
yTrain = labelList
yTest = labelList
# (xTrain, _), (xTest, yTest) = mnist.load_data()
xTrain = xTrain.astype('float32') / 255.
xTest = xTest.astype('float32') / 255.
numTrain = len(xTrain)
numTest = len(xTest)
dimInput = np.prod(xTrain.shape[1:])
sizeDigit = xTrain.shape[1]

xTrain = xTrain.reshape((numTrain, dimInput))
xTest = xTest.reshape((numTest, dimInput))
print(xTrain.shape)
print(xTest.shape)
```

## Construct the encoder and decoder

```python
layEncInt1 = Dense(dimInter1, activation='relu')
layEncInt2 = Dense(dimInter2, activation='relu')
layEncInt3 = Dense(dimInter3, activation='relu')
layEncode = Dense(dimEncode, activation='relu')
layDecInt3 = Dense(dimInter3, activation='relu')
layDecInt2 = Dense(dimInter2, activation='relu')
layDecInt1 = Dense(dimInter1, activation='relu')
layDecode = Dense(dimInput, activation='sigmoid')

inputs = Input(shape=(dimInput,))
interEncode1 = layEncInt1(inputs)
interEncode2 = layEncInt2(interEncode1)
interEncode3 = layEncInt3(interEncode2)
latent = layEncode(interEncode3)
encoder = Model(inputs, latent)

inputLatent = Input(shape=(dimEncode,))
interDecode1 = layDecInt1(inputLatent)
interDecode2 = layDecInt2(interDecode1)
interDecode3 = layDecInt3(interDecode2)
decode = layDecode(interDecode3)
decoder = Model(inputLatent, decode)
```

## Construct the autoencoder

```python
autoencoder = Model(inputs, decoder(encoder(inputs)))
autoencoder.summary()
```

## Compile and train

```python
autoencoder.compile(optimizer=nameOptim, loss='binary_crossentropy')
tic = time()
history = autoencoder.fit(xTrain, xTrain,
                epochs=numEpochs,
                batch_size=sizeBatch,
                shuffle=True,
                validation_data=(xTest, xTest))
timeTrain = time() - tic
```

## Show the historical training progress 

```python
print("traing consumed: " + str(timeTrain) + " seconds")
plotProgress(history)
```

## Get the encoding and decoding results of testing data, and get the mean/std of the encoding 

```python
# encode and decode some digits
# note that we take them from the *test* set
encodeTest = encoder.predict(xTest)
decodeTest = decoder.predict(encodeTest)
meanEncTest = np.mean(encodeTest, axis=0)
stdEncTest = np.std(encodeTest, axis=0)
```

## Compare original digitals with the decoding results

```python
plotCompDecode(xTest, decodeTest, sizeDigit=(100, 140))
```

## Plot the scatter of the encoding space

```python
xlim = (meanEncTest[0] - 3*stdEncTest[0], meanEncTest[0] + 3*stdEncTest[0])
ylim = (meanEncTest[1] - 3*stdEncTest[1], meanEncTest[1] + 3*stdEncTest[1])

plotScatterEncode(encodeTest, yTest, xlim, ylim, numShow=14000, sizeFont=20)
scoreSilh = silhouette_score(encodeTest, yTest)  
```

```python
print(labelSet)
plotScatterEncode(encodeTest, prog, xlim, ylim, sizeFont=1)
```

## Plot the decoding results from the encoding scatter 

```python
# display a 2D manifold of the digits
plotScatterDecode(decoder, (100, 140), xlim, ylim)
```
